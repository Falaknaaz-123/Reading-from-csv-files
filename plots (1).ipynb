```python
#import different libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#read data from the csv file
data = pd.read_csv('Cycle1.csv')
#importing csv file was successful
```

```python
#assigning different columns of the data into different series
current = data['Current']
voltage = data['Voltage']
ah = data['Ah']
wh = data['Wh']
power = data['Power']
battery_temp_degc = data['Battery_Temp_degC']
time = data['Time']
chamber_temp_degc = data['Chamber_Temp_degC']
soc = data['Soc']
timestamp = data['TimeStamp']



```

```python
fig = plt.figure()
myaxes = fig.add_axes([0.1,0.1,5.6,5.6])          #(xmin,ymin,dx,dy) dx is width of the subplot ,dy is the height of subplot
myaxes.plot(time,current,'c',label = 'Current v/s Time')
myaxes.set_xlabel('Time')
myaxes.set_ylabel('Current')
myaxes.set_title('Current v/s Time')
plt.savefig("fig1.pdf",bbox_inches="tight")
```
